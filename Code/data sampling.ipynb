# non probability random sampling

In non-probability random sampling sample is selected in a non-random way. In non-probability random sampling, not all data points have a equal chance of being selected in a sample.

Types of non-probability random sampling are convenience sampling, judgement sampling, quota sampling and snowball sampling.

# convenience sampling
In convenience sampling most accessable data points will be included in sample.

```python
convenience_df = df.head(int(len(df)/10))
```

```python
convenience_df
```

```python
convenience_label=convenience_df['category'].unique()
convenience_label_percentage=[]
for i in range(len(convenience_label)):
    convenience_label_percentage.append(len(convenience_df[convenience_df['category']==convenience_label[i]])*100/len(convenience_df))
```

```python
convenience_label_percentage_df=pd.DataFrame({'name':df_label,'original label percentage':df_label_percentage,
                                  'sample label percentage':[0 for i in range(len(df_label))]})
```

```python
for i in range(len(convenience_label)):
    convenience_label_percentage_df.loc[(convenience_label_percentage_df.name==convenience_label[i]),['sample label percentage']]=convenience_label_percentage[i]
```

```python
convenience_label_percentage_df
```

```python
complete_convenience_df=pd.DataFrame(columns = features)
```

```python
keys=list(convenience_df['pkSeqID'])
```

```python
for i in range(1,75):
    filename=fname+str(i)+extension
    df_temp=pd.read_csv(filename,names=features,low_memory=False)
    df_temp1=df_temp[df_temp['pkSeqID'].isin(keys)]
    complete_convenience_df=complete_convenience_df.append(df_temp1,ignore_index=True)
```

# probability sampling methods
In probability random sampling sample is selected in a random way in which each member of the population has a non-zero probability of getting selected in the sample.
Types of probability random sampling are simple random sampling, systematic sampling, stratified sampling and cluster sampling.

# cluster sampling

```python
interval=1000000
start=[]
stop=[]
for i in range(0,len(df),interval):
    start.append(i)
    if(len(df)>i+1000000):
        stop.append(i+1000000-1)
    else:
        stop.append(len(df)-1)

count=[j-i+1for i,j in zip(start,stop)]
```

```python
cluster=pd.DataFrame({'cluster':['c'+str(i) for i in range(1,75)],'start':start,'stop':stop,'size':count})
cluster
```

```python
sample_cluster=cluster.sample(7)
```

```python
sample_cluster
```

```python
cluster_df=pd.DataFrame(columns = features_1)
```

```python
s_start=list(sample_cluster['start'])
s_stop=list(sample_cluster['stop'])
for i in range(len(s_start)):
    index=np.arange(s_start[i],s_stop[i]+1)
    temp=df.iloc[index]
    cluster_df=cluster_df.append(temp,ignore_index=True)
```

```python
cluster_df
```

```python
cluster_percentage_df=pd.DataFrame({'name':df_label,'original label percentage':df_label_percentage,
                                  'sample label percentage':[0 for i in range(len(df_label))]})
```

```python
cluster_label=cluster_df['category'].unique()
cluster_label_percentage=[]
for i in range(len(cluster_label)):
    cluster_label_percentage.append(len(cluster_df[cluster_df['category']==cluster_label[i]])*100/float(len(cluster_df)))
```

```python
for i in range(len(cluster_label)):
    cluster_percentage_df.loc[(cluster_percentage_df.name==cluster_label[i]),['sample label percentage']]=cluster_label_percentage[i]
```

```python
cluster_percentage_df
```

```python
complete_cluster_df = pd.DataFrame(columns = features)
```

```python
keys=list(cluster_df['pkSeqID'])
```

```python
for i in range(1,75):
    filename=fname+str(i)+extension
    df_temp=pd.read_csv(filename,names=features,low_memory=False)
    df_temp1=df_temp[df_temp['pkSeqID'].isin(keys)]
    complete_cluster_df=complete_cluster_df.append(df_temp1,ignore_index=True)
```

# stratified random sampling

```python
def stratify(dataframe):
    labels=list(dataframe['category'].unique())
    count=[int(len(dataframe[dataframe['category']==i])/10) for i in labels]
    cols=dataframe.columns
    temp=pd.DataFrame(columns =cols)
    for i in range(len(labels)):
        temp=temp.append(dataframe[dataframe['category']==labels[i]].sample(count[i]),ignore_index=True)
    
    return temp
```

```python
stratify_df=stratify(df)
```

```python
stratify_df
```

```python
stratified_percentage_df=pd.DataFrame({'name':df_label,'original label percentage':df_label_percentage,
                                  'sample label percentage':[0 for i in range(len(df_label))]})
```

```python
stratified_label=stratify_df['category'].unique()
stratified_label_percentage=[]
for i in range(len(stratified_label)):
    stratified_label_percentage.append(len(stratify_df[stratify_df['category']==stratified_label[i]])*100/float(len(stratify_df)))
```

```python
for i in range(len(stratified_label)):
    stratified_percentage_df.loc[(stratified_percentage_df.name==stratified_label[i]),['sample label percentage']]=stratified_label_percentage[i]
```

```python
stratified_percentage_df
```

```python
complete_stratified_df = pd.DataFrame(columns = features)
```

```python
keys=list(stratify_df['pkSeqID'])
```

```python
for i in range(1,75):
    filename=fname+str(i)+extension
    df_temp=pd.read_csv(filename,names=features,low_memory=False)
    df_temp1=df_temp[df_temp['pkSeqID'].isin(keys)]
    del(df_temp)
    complete_stratified_df=complete_stratified_df.append(df_temp1,ignore_index=True)
```
